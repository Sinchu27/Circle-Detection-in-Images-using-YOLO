CREATING THE DATASET CONTAINING IMAGES WITH CIRCLES

```python
import cv2
import numpy as np
import os
import random

# Parameters
num_images = 100        # Total number of images to generate
image_size = (416, 416) # Size of the images (width, height)
min_radius = 10         # Minimum circle radius
max_radius = 50         # Maximum circle radius
output_dir = 'circle_images'  # Directory to save images and annotations

# Create output directory
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Function to generate random color
def random_color():
    return tuple([random.randint(0, 255) for _ in range(3)])

# Function to generate a random background
def random_background():
    return np.full((image_size[1], image_size[0], 3), random.randint(200, 255), dtype=np.uint8)

# Generate images
for i in range(num_images):
    # Create a blank background
    background = random_background()

    # Determine if the image will have a circle or not
    has_circle = random.choice([True, False])

    if has_circle:
        # Generate random circle properties
        radius = random.randint(min_radius, max_radius)
        center = (random.randint(radius, image_size[0] - radius), random.randint(radius, image_size[1] - radius))
        color = random_color()

        # Draw the circle on the image
        cv2.circle(background, center, radius, color, -1)

        # Save annotation in YOLO format
        x_center = center[0] / image_size[0]
        y_center = center[1] / image_size[1]
        width = height = (2 * radius) / image_size[0]

        annotation = f"0 {x_center} {y_center} {width} {height}"
        annotation_path = os.path.join(output_dir, f'image_{i:03d}.txt')
        with open(annotation_path, 'w') as f:
            f.write(annotation)

    # Save the image
    image_path = os.path.join(output_dir, f'image_{i:03d}.jpg')
    cv2.imwrite(image_path, background)
```

```python
import os
import shutil
import random

# Paths
source_dir = '/content/train_images'  # Original dataset directory
val_dir = '/content/val_images'       # Validation set directory
test_dir = '/content/test_images'     # Test set directory

# Create directories if they don't exist
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Get list of all image files
all_images = os.listdir(source_dir)

# Shuffle the images
random.shuffle(all_images)

# Define split ratios
val_ratio = 0.2  # 20% for validation
test_ratio = 0.1 # 10% for testing

# Calculate the number of images for validation and test
val_count = int(len(all_images) * val_ratio)
test_count = int(len(all_images) * test_ratio)

# Move files to validation and test directories
val_images = all_images[:val_count]
test_images = all_images[val_count:val_count + test_count]

for img in val_images:
    shutil.move(os.path.join(source_dir, img), os.path.join(val_dir, img))

for img in test_images:
    shutil.move(os.path.join(source_dir, img), os.path.join(test_dir, img))

# Verify the move
print(f"Training set: {len(os.listdir(source_dir))} images")
print(f"Validation set: {len(os.listdir(val_dir))} images")
print(f"Test set: {len(os.listdir(test_dir))} images")
```

```python
# Define the YAML content
yaml_content = """
train: train_images  # Path to the training images directory
val: val_images      # Path to the validation images directory
test: test_images    # Path to the test images directory

nc: 1                # Number of classes
names: ['circle']    # Class names
"""

# Save the YAML content to a file
with open('your_dataset.yaml', 'w') as f:
    f.write(yaml_content)
```
